```python
import xlwings as xw
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import os

# 定義繪圖函數
def Pareto(problems, values, output_file):
    df = pd.DataFrame({'Values': values})
    df.index = problems
    df = df.sort_values(by='Values', ascending=False)
    df["cumpercentage"] = df["Values"].cumsum() / df["Values"].sum() * 100
    plt.figure(figsize=(10, 6))
    plt.bar(df.index, df["Values"], color="C0")
    plt.xticks(rotation=90)
    ax = plt.twinx()
    ax.plot(df.index, df["cumpercentage"], color="C9", marker=".")
    ax.yaxis.set_major_formatter(PercentFormatter())
    plt.title("Pareto Diagram")
    plt.tick_params(axis="x", colors="green")
    plt.tick_params(axis="y", colors="k")
    plt.grid()
    plt.savefig(output_file)
    plt.close()

def generate_pareto_chart():
    file_path = 'D:/Desktop/Book1.xlsx'  # Excel 文件路徑
    chart_file = 'D:/Desktop/pareto_chart.png'  # 圖像文件路徑

    # 檢查文件存在性
    if not os.path.isfile(file_path):
        raise FileNotFoundError(f"The file was not found: {file_path}")

    # 打開 Excel 文件
    wb = xw.Book(file_path)
    sheet = wb.sheets['Sheet1']

    # 讀取數據
    df = sheet.range('A1:B13').options(pd.DataFrame, index=False, header=True).value
    
    print("列名:", df.columns.tolist())
    
    if 'Problems' not in df.columns or 'Values' not in df.columns:
        raise ValueError("數據框中缺少 'Problems' 或 'Values' 列")
    
    problems = df['Problems']
    values = df['Values']
    
    # 繪製 Pareto 圖
    Pareto(problems, values, chart_file)
    
    # 檢查圖像文件是否成功創建
    if not os.path.isfile(chart_file):
        raise FileNotFoundError(f"The chart file was not found: {chart_file}")

    # 將圖表插入 Excel
    sheet.pictures.add(chart_file, name='Pareto Chart', update=True, left=sheet.range('A1:B13').left, top=sheet.range('A1:B13').top, width=600, height=400)
    print("Pareto chart inserted into Excel.")

if __name__ == "__main__":
    generate_pareto_chart()
```

```python
import xlwings as xw
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import os

# 定義繪圖函數
def Pareto(problems, values, output_file):
    df = pd.DataFrame({'Values': values})
    df.index = problems
    df = df.sort_values(by='Values', ascending=False)
    df["cumpercentage"] = df["Values"].cumsum() / df["Values"].sum() * 100
    plt.figure(figsize=(10, 6))
    plt.bar(df.index, df["Values"], color="C0")
    plt.xticks(rotation=90)
    ax = plt.twinx()
    ax.plot(df.index, df["cumpercentage"], color="C9", marker=".")
    ax.yaxis.set_major_formatter(PercentFormatter())
    plt.title("Pareto Diagram")
    plt.tick_params(axis="x", colors="green")
    plt.tick_params(axis="y", colors="k")
    plt.grid()
    plt.savefig(output_file)
    plt.close()

def generate_pareto_chart():
    file_path = 'D:/Desktop/Book1.xlsx'  # Excel 文件路徑
    chart_file = 'D:/Desktop/pareto_chart.png'  # 圖像文件路徑

    # 檢查文件存在性
    if not os.path.isfile(file_path):
        raise FileNotFoundError(f"The file was not found: {file_path}")

    # 打開 Excel 文件
    wb = xw.Book(file_path)
    sheet = wb.sheets['Sheet1']

    # 讀取數據
    df = sheet.range('A1:B13').options(pd.DataFrame, index=False, header=True).value
    
    print("列名:", df.columns.tolist())
    
    if 'Problems' not in df.columns or 'Values' not in df.columns:
        raise ValueError("數據框中缺少 'Problems' 或 'Values' 列")
    
    problems = df['Problems']
    values = df['Values']
    
    # 繪製 Pareto 圖
    Pareto(problems, values, chart_file)
    
    # 檢查圖像文件是否成功創建
    if not os.path.isfile(chart_file):
        raise FileNotFoundError(f"The chart file was not found: {chart_file}")

    # 將圖表插入 Excel
    # 確定插入位置，這裡使用 'A1' 的位置
    sheet.pictures.add(chart_file, name='Pareto Chart', update=True, left=sheet.range('A1').left, top=sheet.range('A1').top, width=600, height=400)
    print("Pareto chart inserted into Excel.")

if __name__ == "__main__":
    generate_pareto_chart()
```
